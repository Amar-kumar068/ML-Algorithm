```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
data=load_digits()
```

```python
dir(data)
```

```python
plt.gray()
for i in  range(4):
    plt.matshow(data.images[i])
```

```python
data.target[:5]
```

```python
df=pd.DataFrame(data.data)
```

```python
df
```

```python
df['Target']=data.target
```

```python
df
```

```python
y=df['Target']
```

```python
X=df.drop('Target',axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
X_test
```

```python
pred
```

```python
## Confusion Matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## IRIS DataSet Prediction

```python
from sklearn.datasets import load_iris
idata=load_iris()
```

```python
dir(idata)
```

```python
df1=pd.DataFrame(idata.data,columns=idata.feature_names)
```

```python
df1
```

```python
df1['target']=idata.target
```

```python
df1
```

```python
Xd=df1.drop('target',axis=1)
```

```python
Xd
```

```python
yd=df1.target
```

```python
yd
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xd, yd, test_size=0.33, random_state=42)
```

```python
model1=RandomForestClassifier(n_estimators=20)
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
pred1=model1.predict(X_test)   
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred1))
```
